```python
##DEFINICION DE LAS VARIABLES EMPLEDAS

    #GLOBALES
M = 0 #Mach de vuelo
Cp = 0 # J/(kgK)
g = 9.81 #m/s^
Ra = 287 #J/kg/k
gamma = 0 #
Mc = 12 #g/mol Masa molecular carbono
Mh = 1 #g/mol Masa molecular hidrogeno
Mn = 14 #g/mol Masa molecular nitrogeno
Mo = 16 #g/mol Masa molecular oxigeno
Mair = 0.79*2*Mn + 0.21*2*Mo #g/mol Masa del aire
Mcom = (0.25*(8*Mc + 18*Mh)) + (0.25*(7*Mc + 16*Mh)) + (0.5*(14*Mc + 30*Mh)) 
    
    #DIFUSOR (di) (1-2)
re_di = 1 #Rendimiento adiabatico del Di
G1_ = 0 #Gasto sin corregir
G1 = 0 #Gasto a la entrada del di corregido, a partir de ahora todos estan corregidos
G2 = 0 #Gasto a la salida del di/entrada Fan
T1t = 0 #Temperatura de remanso a la entrada del di
T2t = 0 #Temperatura de remanso a la salida del di/entrada Fan
T1_ = 0 #Temperatura estatica sin corregir
T1 = 0 #Temperatura estatica a la entrada del di corregida
T2 = 0 #Temperatura estatica a la salida del di/entrada Fan
p1t = 0 #Presion de remanso a la entrada del di
p2t = 0 #Presion de remanso a la salida del di/entrada Fan
p1_ = 0 #Presion estatica sin corregir
p1 = 0 #Presion estatica a la entrada del di corregida
p2 = 0 #Presion estatica a la salida del di/entrada Fan
v1 = 0 #Velocidad entrada Difusor/velocidad de vuelo
v2 = 0 #Velocidad salida Difusor/entrada Fan
A1 = 0 #Area entrada Difusor
M_di = 0 #Mach en eldifusor

    #Fan (2-25)(12-13)
pi_fan = 0 #Relacion de compresion variable
re_fan = 0 #Rendimiento adiabatico variable
G2 = 0 #Gasto entrante al fan que pasa por la zona primaria
G12 = 0 #Gasto entrante al fan que pasa por la zona secundaria
G25 = 0 #Gasto saliente del fan que pasa por la zona primaria/entrante al compresor de baja
G13 = 0 #Gasto saliente del fan que pasa por la zona secundaria/entrante a la tobera secundaria
T2t = 0 #Temperatura de remanso a la entrada del fan zona primaria
T25t = 0 #Temperatura de remanso a la salida del fan zona primaria/entrante al compresor de baja
T12t = 0 #Temperatura de remanso a la entrada del fan zona secundaria
T13t = 0 #Temperatura de remanso a la salida del fan zona secundaria/entrante a la tobera secundaria
p2t = 0 #Presion de remanso a la entrada del fan zona primaria
p25t = 0 #Presion de remanso a la salida del fan zona primaria/entrante al compresor de baja
p12t = 0 #Presion de remanso a la entrada del fan zona secundaria
p13t = 0 #Presion de remanso a la salida del fan zona secundaria/entrante a la tobera secundaria
T2 = 0 #Temperatura estatica a la entrada del fan zona primaria
T25 = 0 #Temperatura estatica a la salida del fan zona primaria/entrante al compresor de baja
T12 = 0 #Temperatura estatica a la entrada del fan zona secundaria
T13 = 0 #Temperatura estatica a la salida del fan zona secundaria/entrante a la tobera secundaria
p2t = 0 #Presion estatica a la entrada del fan zona primaria
p25t = 0 #Presion estatica a la salida del fan zona primaria/entrante al compresor de baja
p12t = 0 #Presion estatica a la entrada del fan zona secundaria
p13t = 0 #Presion estatica a la salida del fan zona secundaria/entrante a la tobera secundaria
v2 = 0 #Velocidad entrante al fan zona primaria
v12 = 0 #Velocidad entrante al fan zona secundaria
v25 = 0 #Velocidad saliente del fan zona primaria/entrante compresor de baja
v13 = 0 #Velocidad saliente del fan zona secundaria/entrante tobera secundaria
D_fan = 2.5 #Diametro del fan en metros
M_fan = 0 #Mach en el fan

    #COMPRESOR DE BAJA (cb) (25-26)
re_cb = 0.89 #Rendimiento adiabatico del cb    
pi_cb = 2.5 #Relacion de compresion del cb
G26 = 0 #Gasto a la salida del cb/entrada Jet-Pipe1
T26t = 0 #Temperatura de remanso a la salida del cb/entrada Jet-Pipe1
T26 = 0 #Temperatura estatica a la salida del cb/entrada Jet-Pipe1
p26t = 0 #Presion de remanso a la salida del cb/entrada Jet-Pipe1
p26 = 0 #Presion estatica a la salida del cb/entrada Jet-Pipe1
W_cb = 0 #Trabajo del cb

    #JET-PIPE1 (jp1) (26-27)
Gs = 3 #Gasto sangrado
G27 = 0 #Gasto salida Jp1/entrada Compresor de Alta
T27t = 0 #Temperatura de remanso a la salida del Jp1/entrada Compresor de Alta
T27 = 0 #Temperatura estatica a la salida del Jp1/entrada Compresor de Alta
p27t = 0 #Presion de remanso a la salida del Jp1/entrada Compresor de Alta
p27 = 0 #Presion estatica a la salida del Jp1/entrada Compresor de Alta

    #COMPRESOR DE ALTA (ca) (27-3)
re_ca = 0 #Rendimiento adiabatico del ca
pi_ca = 0 #Relacion de compresion del ca
G3 = 0 #Gasto a la salida del ca/entrada Camara de Combustion
T3t = 0 #Temperatura de remanso a la salida del ca/entrada Camara de Combustion
T3 = 0 #Temperatura estatica a la salida del ca/entrada Camara de Combustion
p3t = 0 #Presion de remanso a la salida del ca/entrada Camara de Combustion
p3 = 0 #Presion estatica a la salida del ca/entrada Camara de Combustion
W_ca = 0 #Trabajo del ca

    #CAMARA DE COMBUSTION (cc) (3-4)
re_cc = 0.94 #Rendimiento de la combustion
G4 = 0 #Gasto a la salida de cc/entrada Turbina de Alta
c1 = 0 #Gasto de combustible
T4t = 0 #Temperatura de remanso a la salida de cc/entrada Turbina de Alta
T4 = 0 #Temperatura estatica a la salida de cca/entrada Turbina de Alta
p4t = 0 #Presion de remanso a la salida de cc/entrada Turbina de Alta
p4 = 0 #Presion estatica a la salida de cc/entrada Turbina de Alta
Q_c1 = 0 #Calor de la combustion
L = 44.5*10**6 # J/Kg Poder calorifico del combustible
f1 = 0 #Dosado
fs1 = 0 #Dosado estequeometrico
Mc = 152.5 #g/mol Masa molecular del combustible

    #TURBINA DE ALTA (ta) (4-45)
re_ta = 0.86 #Rendimiento adiabatico de la ta
G_ta = 0 #Gasto ta
T45t = 0 #Temperatura de remanso a la salida de la ta/entrada Jet-Pipe2
T45 = 0 #Temperatura estatica a la salida de la ta/entrada Jet-Pipe2
p45t = 0 #Presion de remanso a la salida de la ta/entrada Jet-Pipe2
p45 = 0 #Presion estatica a la salida de la ta/entrada Jet-Pipe2
W_ta = 0 #Trabajo de la ta

    #TURBINA DE BAJA (tb) (45-5)
re_tb = 0.82 #Rendimiento adiabatico de la tb
G_tb = 0 #Gasto tb
T5t = 0 #Temperatura de remanso a la salida de la tb/entrada Postcombustor
T5 = 0 #Temperatura estatica a la salida de la tb/entrada Postcombustor
p5t = 0 #Presion de remanso a la salida de la tb/entrada Postcombustor
p5 = 0 #Presion estatica a la salida de la tb/entrada Postcombustor
W_tb = 0 #Trabajo de la tb

    #TOBERA PRINCIPAL (top) (7-8)
re_top = 1 #Rendimiento adiabatico de la top
G8 = 0 #Gasto salida top
T8t = 0 #Temperatura de remanso a la salida de la top
T8 = 1300 #Temperatura estatica a la salida de la top
p8t = 0 #Presion de remanso a la salida de la top
p8 = 0 #Presion estatica a la salida de la top
v8 = 0 #Velocidad salida top   
A8 = 0 #Area salida top
M_top = 0 #Mach en la tobera principal
    
    #TOBERA SECUNDARIA (tos) (17-18)
re_tos = 0.92 #Rendimiento adiabatico de la tos
G17 = 0 #Gasto entrada tos
G18 = 0 #Gasto salida tos
T17t = 0 #Temperatura de remanso a la salida de la jp2/entrada tos
T17 = 0 #Temperatura estatica a la salida de la jp2/entrada tos
p17t = 0 #Presion de remanso a la salida de la jp2/entrada tos
p17 = 0 #Presion estatica a la salida de la jp2/entrada tos
T18t = 0 #Temperatura de remanso a la salida de la tos
T18 = 0 #Temperatura estatica a la salida de la tos defino una teperatura inicial para no tener que iterar, este valor es para el cual M_tos<=1
p18t = 0 #Presion de remanso a la salida de la tos
p18 = 0 #Presion estatica a la salida de la tos   

##CALCULOS INICIALES
Tsl = 288.15 #K = 15ºc
psl = 101325 # Pa
alpha = 6.5*1e-3 #ºC/m

def temp_isa(h):
    T = Tsl - alpha*h #h metros
    return T

def cp(T):
    if T < 1000:
        a0 = 3.56839620
        a1 = -6.78729429*10**-4
        a2 = 1.55371476*10**-6
        a3 = -3.29937060*10**-12
        a4 = -4.66395387*10**-13 
        cp= Ra*(a0 + a1*T + a2*T**2 + a3*T**3 + a4*T**4)
    if T >= 1000:
        a0 = 3.08792717
        a1 = 1.24597184*10**-3
        a2 = -4.23718945*10**-7
        a3 = 6.74774789*10**-11
        a4 = -3.97076972*10**-15 
        cp= Ra*(a0 + a1*T + a2*T**2 + a3*T**3 + a4*T**4)
    return cp

def gamma(T):
    gamma = 1.405-(0.11*(T-175)/1825)
    return gamma

def pres_isa(h):
    a = (1-alpha*h/Tsl)**(g/alpha/Ra)
    p = psl*a
    return p
```

```python
###RESOLUCION PARA h=0
#Estas listas vacias son para poder usar pandas
lista1=[]
lista2=[]
lista3=[]
lista4=[]
lista5=[]
lista6=[]
lista7=[]
lista8=[]
lista9=[]
lista10=[]
lista11=[]
lista12=[]
lista13=[]
lista14=[]
lista15=[]
lista16=[]
lista17=[]
lista18=[]
lista19=[]
lista20=[]
lista21=[]
lista22=[]
lista23=[]
lista24=[]
lista25=[]
lista26=[]
lista27=[]
lista28=[]

h = 0 #h en metros
T0 = temp_isa(h)
T1 = T0 #Sin corregir
p0 = pres_isa(h)
p1 = p0 #Sin corregir
vel1 = 95 #Velocidad minima en nudos
vel2 = 501 #Velocidad maxima en nudos+1 para poder coger el 500
D_tr = 0.715
A_tr = np.pi*(D_tr**2)/4 #Area interna del generador de gas
p8 = p0
for v in range (vel1,vel2,15):
    v1 = v*1852/3600
    ro = p0/(Ra*T0)
    D_efectivo = 2*(1.70*D_fan/3.85)
    Mv = v1/(np.sqrt(gamma(T1)*Ra*T1))
    A_fan = np.pi*((D_efectivo**2)/4)
    G_fan = ro*A_fan*v1
    
    
    ##DIFUSOR (di)    
    T1t = T1 + ((v1**2/(2*cp(T1))))
    p1 = p11 = p0 
    p1t = ((1+(((gamma(T1t)-1)*(Mv**2)))/2)**(gamma(T1t)/(gamma(T1t)-1)))*p1
    p2t = p1t
    T2t = T1t
    T12t = T1t #12 es la entrada al fan por la parte "secundaria", aunque aun no hay division fisica
    p12t = p2t #2 es la entrada al fan por la parte "primaria", aunque aun no hay division fisica
    M_di = np.sqrt(((T1t/T1)-1)*(2/(gamma(T1t)-1)))
    lista1.append (f"{v1:.2f}")
    lista2.append (f"{G_fan:.2f}")
              
    ##Fan (fan)
    G12 = G2 = G_fan
    G_pi = ro *A_tr* v1 #pi corresponde al gasto primario
    G_sig = G_fan - G_pi #sig corresponde al gasto secundario
    lam = G_sig/G_pi
    A_di = G_pi*(lam+1)/(ro*v1)
    G1 = ro*A_di*v1
    
    def  pi_fan(G):
        if 0 < G < 50:
            pi_fan = (1.5*G+50)/50
            return pi_fan
        if 50 < G < 100:
            pi_fan = (G-50+250)/100
            return pi_fan
        if 100 < G < 400:
            pi_fan = ((G-100)/300)*0.25 + 3 
            return pi_fan
        if G > 400:
            pi_fan = 3.2
            return pi_fan
    
    def  re_fan(G):   
        if 0 < G < 50:
            re_fan = 0.75
            return re_fan
        if 50 < G < 100:
            re_fan = 0.85
            return re_fan
        if 100 < G < 400:
            re_fan = 0.72
            return re_fan 
        if G > 400:
            re_fan = 0.72
            return re_fan
        
    T25t_id = ((((((pi_fan(G_fan)))**((1004-1)/1004))-1)/(re_fan(G_fan)))+1)*T2t #Se usa un T25t ideal para poder hacer la media
    T25t = (((((pi_fan(G_fan))**((gamma((T25t_id+T2t)/2)-1)/gamma((T25t_id+T2t)/2)))-1)/(re_fan(G_fan)))+1)*T2t
    p25t = p2t*(pi_fan(G_fan))
    G25 = G_pi 
    W_fan = G25*cp(T25t)*T25t - G2*cp(T2t)*T2t        
        
    T13t = T25t
    p13t = p25t
    G13 = G_sig       
    tau_fan = cp((T13t+T12t)/2)*(T13t-T12t)
        
    lista3.append (f"{G_pi:.2f}")
    lista4.append (f"{G_sig:.2f}")
    lista5.append (f"{lam:.4f}")
    lista6.append (f"{A_di:.6f}")
    lista7.append (f"{pi_fan(G_fan):.4f}")
    lista8.append (f"{re_fan(G_fan):.4f}") 
    
        
    ##COMPRESOR DE BAJA (cb)
    T26t_id = (((((pi_cb)**((1004-1)/1004))-1)/re_cb)+1)*T25t #Se usa un T25t ideal para poder hacer la media
    T26t = (((((pi_cb)**((gamma((T26t_id+T25t)/2)-1)/gamma((T26t_id+T25t)/2)))-1)/re_cb)+1)*T25t
    p26t = p25t*pi_cb
    G26 = G25 
    W_cb = G26*cp(T26t)*T26t - G25*cp(T25t)*T25t
    
             
    ##JET-PIPE1 (jp_1)
    G27 = G26 - Gs
    T27t = T26t
    p27t = 0.98*p26t
    
                
    ##COMPRESOR DE ALTA (ca)
    re_ca = 0.95 - (0.2*Mv) 
    pi_ca = 2 + 5*Mv 
    T3t_id = (((((pi_ca)**((1004-1)/1004))-1)/re_ca)+1)*T27t #Se usa un T3t ideal para poder hacer la media
    T3t = (((((pi_ca)**((gamma((T3t_id+T27t)/2)-1)/gamma((T3t_id+T27t)/2)))-1)/re_ca)+1)*T27t 
    p3t = p27t*pi_ca
    G3 = G27 
    W_ca = G3*cp(T3t)*T3t - G27*cp(T27t)*T27t
    
        
    ##CAMARA DE COMBUSTION
    p4t = p3t*0.95
    fs = Mc/(4.76*(Mair)*12.5)
    f = 0.53*fs
    c = f*G3
    G4 = G3 + c
    T4t_id = ((re_cc*c*L) + (G3*1004*T3t))/(G4*1004)
    T4t = ((re_cc * c* L) + (G3 * cp((T4t_id+T3t)/2) * T3t))/(G4*cp((T4t_id+T3t)/2))
    
         
    ##TURBINA DE ALTA (ta)
    T45t = T4t + T27t - T3t #Acople mecanico compresor-turbina
    p45t = (((((T45t/T4t)-1)/re_ta)+1)**(gamma((T45t+T4t)/2)/(gamma((T45t+T4t)/2)-1)))*p4t
    G45 = G4
    
        
    ##TURBINA DE BAJA (tb)    
    T5t = T45t - T25t + T1t
    p5t = (((((T5t/T45t)-1)/re_tb)+1)**(gamma((T5t+T45t)/2)/(gamma((T5t+T45t)/2)-1)))*p45t 
    G5 = G45

    
    ##TOBERA PRINCIPAL (top) #no interesa que empuje, se busca que sea critica
    
    G7 = G5
    G8 = G7
    T7t = T5t
    T8t = T7t
    p7t = p5t
    p8t = p7t
    
    T8 = T8t*(2/(gamma(T8t)+1)) #Formula para M=1 
    v8 = np.sqrt((T8t-T8)*2*(((cp(T8t)+cp(T8))/2)))
    p8 = (((((T8/T8t)-1)/re_top)+1)**(gamma(T8t)/(gamma(T8t)-1)))*p8t 
    ro8 = p8/(Ra*T8)
    A8 = G8/(ro8*v8)
    
    M_top = np.sqrt(((T8t/T8)-1)*(2/(gamma(T8t)-1)))         
    E_top = G8*v8 - G_pi*v1 + A8*(p8-p1)
    D_top = np.sqrt(A8*4/np.pi) #Diametro (interno) del generador de gas   

    if T8t > 1950:
        lista9.append ("Ex")
    if T8t < 1950:
        lista9.append (f"{T8t:.2f}")
      
    if 0.15*A_di< A8 < 0.75*A_di:
        lista10.append (A8)
    else:
        lista10.append (f"{A8:.6f}")
    

    lista11.append (f"{M_top:.6f}")
          
    lista12.append (f"{E_top:.1f}")  
    lista27.append (f"{p8t:.2f}")
    lista28.append (f"{p8:.2f}")
    
    ##TOBERA SECUNDAR IA (tos) Se busca Mach critico
    G17 = G13 
    G18 = G17
    T17t= T13t
    T18t = T17t
    p17t = p13t 
    p18t = p17t 
    
    #Para la resolucion de esta parte se busca que sea critica
    T18 = T18t*(2/(gamma(T18t)+1)) #Formula para M=1
    p18 = (((((T18/T18t)-1)/re_tos)+1)**(gamma(T18t)/(gamma(T18t)-1)))*p18t 
    v18 = np.sqrt((T18t-T18)*2*cp(T18t))
    ro18 = p18/(Ra*T18)
    A18 = G18/(ro18*v18)
    M_tos = np.sqrt(((T18t/T18)-1)*(2/(gamma(T18t)-1)))         
    E_tos = G18*v18 - G_sig*v1 + A18*(p18-p11)
        
    if T18t > 1950:
        lista16.append ("Ex")
    if T18t < 1950:
        lista16.append (f"{T18t:.2f}")   
        
    cs = (c)/(E_tos+E_top) #Consumo especifico   
     
    re8_motor = ((G8)*(v8**2) - G_pi*(v1**2))/(2*c*L)
    re8_prop = (2*v1*((G8)*v8 - G_pi*v1))/((G8)*(v8**2) - G_pi*(v1**2))
    re8_motor_prop = (v1*((G8)*v8 - G_pi*v1))/(c*L)
    
    re18_motor = ((G18)*(v18**2) - G_sig*(v1**2))/(2*(c)*L)
    re18_prop = (2*v1*((G18)*v18 - G_sig*v1))/((G18)*(v18**2) - G_sig*(v1**2))
    re18_motor_prop = (v1*((G18)*v18 - (G_sig*v1)))/((c)*L)
    
    re_motor = (G_pi/G_fan)*re8_motor + (G_sig/G_fan)*re18_motor 
    re_prop = (G_pi/G_fan)*re8_prop + (G_sig/G_fan)*re18_prop
    re_motor_prop = (G_pi/G_fan)*re8_motor_prop + (G_sig/G_fan)*re18_motor_prop
    E = E_top+E_tos 
    
    lista18.append (f"{M_tos:.1f}")   
    lista17.append (f"{A18:.6f}") 
    lista19.append (f"{E_tos:.1f}")   
    lista20.append (f"{cs:.8f}")
    lista24.append (f"{re_motor:.3f}")
    lista25.append (f"{re_prop:.3f}")
    lista26.append (f"{re_motor_prop:.3f}")
    lista13.append (f"{E:.1f}")
    
    
    diccionario = {"velocidad (m/s)" :  lista1 , "M_top" : lista11 , "M_tos" : lista18 , "E_top (N)" : lista12 ,
                   "E_tos (N)" : lista19 , "E (N)" : lista13 , "cs" : lista20 , "G_fan (Kg/s)" : lista2 , "G_pi (Kg/s)" : lista3 , 
                   "G_sig (Kg/s)" : lista4 , "Lambda" : lista5 , "A_di (m^2)" : lista6 , "Pi_fan" : lista7 , 
                   "re_fan" : lista8 , "T8t (K)" : lista9 , "A8 (m^2)" : lista10 , "p8t (pa)" : lista27 , 
                   "p8 (pa)" : lista28 , "T18t (K)" : lista16 , "A18 (m^2)" : lista17 , "re_motor" : lista24 ,
                   "re_prop" : lista25 , "re_motor_prop" : lista26}
    
    
    df = pd.DataFrame.from_dict(diccionario)
```
